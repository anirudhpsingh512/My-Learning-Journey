```python
class Animal:
  def __init__(self, name, species):
    self.name = name
    self.species = species
  def __init__(self, name, species,age):
    self.name = name
    self.species = species
    self.age = age

  def make_sound(sound):
    print(f"{self.name} the {self.species} says {sound}")

#dog = Animal("dog", "mammals")   #Error because of constructor overrinding
```

```python
#To fix this -
class Animal:
  def __init__(self, *args):
    if len(args) == 1:
      self.name = args[0]
    elif len(args) == 2:
      self.name = args[0]
      self.species = args[1]
    else:
      self.name = args[0]
      self.species = args[1]
      self.age = args[2]

dog = Animal("dog", "mammals")
chicken = Animal("chicken", "bird", 2)
print(dog.name, dog.species)
print(chicken.name, chicken.species, chicken.age)
```

#Encapsulation

Private - variables and methods can only be used in the same class <br>
Protected - variables and methods can only be used in the derived classed <br>
Public - variables and methods can be used anywhere

```python
class Car:
    def __init__(self, color):
        self.color = color  # Public attribute

    def start(self):
        print("Car started.")

my_car = Car("Red")
print(my_car.color)  # Accessing public attribute
my_car.start()
```

```python
class Car:
    def __init__(self, color):
        self.__color = color  # Private attribute

    def get_color(self):
        return self.__color

my_car = Car("Blue")
# print(my_car.__color)  # Direct access (usually discouraged)
print(my_car.get_color())  # Accessing through a public method
```

```python
class Vehicle:
    def __init__(self, max_speed):
        self._max_speed = max_speed  # Protected attribute

class Car(Vehicle):
    def __init__(self, color, max_speed):
        super().__init__(max_speed)
        self.color = color

    def get_max_speed(self):
        return self._max_speed

my_car = Car("Red", 180)
print(my_car.get_max_speed())
```
